```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import pygwalker as pyg
```

```python
# Load the data
data = pd.read_csv('Coursera_Cleaned_v1.csv')
```

```python
walker = pyg.walk(data)
```

```python
# for every categorical
# for every numerical
# show barplot

# Get the categorical columns
categorical_columns = ['Category', 'Sub-Category', 'Course Type', 'Language']

# Get the numerical columns
numerical_columns = ['Rating', 'Number of viewers', 'Duration']

# Nested for loop to iterate over categorical and numerical columns
for categorical_column in categorical_columns:
    # Add count of records barplot
    sns.countplot(x=categorical_column, data=data)
    plt.xticks(rotation=90)  # Rotate x-axis labels vertically
    plt.title(f"Count of courses by {categorical_column}")
    plt.show()
    
    for numerical_column in numerical_columns:
        # Show barplot
        sns.barplot(x=categorical_column, y=numerical_column, data=data)
        plt.xticks(rotation=90)  # Rotate x-axis labels vertically
        plt.title(f"Barplot of {numerical_column} by {categorical_column}")
        plt.show()
        
        # Add sum of views barplot
        if numerical_column == 'Number of viewers':
            sns.barplot(x=categorical_column, y=numerical_column, data=data.dropna(), estimator=sum, ci=None)
            plt.xticks(rotation=90)  # Rotate x-axis labels vertically
            plt.title(f"Sum of {numerical_column} by {categorical_column}")
            plt.show()

```
